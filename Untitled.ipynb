## Question 2: Extracting Tesla Revenue Data Using Webscraping 

```python
!pip install requests beautifulsoup4 pandas
```

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"


response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find_all("table")[1]  

rows = table.find_all("tr")[1:]

dates = []
revenues = []

for row in rows:
    cols = row.find_all("td")
    if len(cols) == 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip().replace('$', '').replace(',', '')
        if revenue != "":
            dates.append(date)
            revenues.append(float(revenue))

df = pd.DataFrame({'Date': dates,'Revenue (in millions)': revenues})

df.tail()
```

# Question4: Extracting GameStop Stock Data Using Webscraping 

```python
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(response.text, 'html.parser')

tables = soup.find_all("table")
for table in tables:
    if "GameStop Quarterly Revenue" in str(table):
        revenue_table = table
        break

data = []
for row in revenue_table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) == 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip().replace('$', '').replace(',', '')
        if revenue != "":
            data.append((date, float(revenue)))

gme_revenue = pd.DataFrame(data, columns=["Date", "Revenue"])

gme_revenue.tail()
```

## Question 5: Tesla Stock and Revenue Dashboard

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt 
```

```python
tesla_stock_data = {'Date': ['Jan', 'Feb', 'Mar', 'Apr'],'Close': [240, 250, 265, 258]}
tesla_revenue_data = {'Date': ['Q1', 'Q2', 'Q3', 'Q4'],'Revenue': [21000, 23000, 22000, 25000] }

def make_graph(x, y, title, xlabel, ylabel):
    plt.figure(figsize=(6,4))
    plt.plot(x, y, marker='o')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
make_graph(tesla_stock_data['Date'], tesla_stock_data['Close'],"Tesla Stock Price Over Time", "Month", "Price (USD)")
make_graph(tesla_revenue_data['Date'], tesla_revenue_data['Revenue'],"Tesla Revenue Per Quarter", "Quarter", "Revenue (Millions USD)")
```

## Question 6: Gamestop Stock and Revenue Dashboard 

```python
!pip install matplotlib
!pip install pandas
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
stock_data = {'Date': ['2025-04-01', '2025-04-05', '2025-04-10', '2025-04-15', '2025-04-17'],'Stock Price': [23.45, 24.10, 25.35, 26.10, 26.78]}
revenue_data = {'Fiscal Year': ['2020', '2021', '2022', '2023', '2024'],'Revenue (Billion USD)': [6.47, 5.09, 6.01, 5.27, 3.82]}

stock_df = pd.DataFrame(stock_data)
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
revenue_df = pd.DataFrame(revenue_data)

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(stock_df['Date'], stock_df['Stock Price'], marker='o', color='blue')
plt.title('GameStop Stock Price - April 2025')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.subplot(1, 2, 2)
plt.bar(revenue_df['Fiscal Year'], revenue_df['Revenue (Billion USD)'], color='green')
plt.title('GameStop Annual Revenue (FY 2020 - FY 2024)')
plt.xlabel('Fiscal Year')
plt.ylabel('Revenue (Billion USD)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
